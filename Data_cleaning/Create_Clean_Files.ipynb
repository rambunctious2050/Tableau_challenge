```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import datetime
```

# Change the file path, and run the following code for each month

```python
# Files to load
file = "../Citibike_data/Feb_Mar_Apr_Data/202004-citibike-tripdata.csv"
```

```python
# Read file and store into Pandas data frame
file1 = pd.read_csv(file)
```

```python
df = pd.DataFrame(file1)
```

```python
df.head()
```

```python
# Use "starttime" to extract the year, month, and day for the trip
trip_year = pd.DatetimeIndex(df["starttime"]).year
trip_month = pd.DatetimeIndex(df["starttime"]).month
trip_day = pd.DatetimeIndex(df["starttime"]).day

# Add trip year, month and day to the dataset
df.insert(15,"Trip_Year",trip_year)
df.insert(16,"Trip_Month",trip_month)
df.insert(17,"Trip_Day",trip_day)
```

```python
df.head()
```

```python
# FILE 1 - Contains count and average trip duration per day. Generate one for each month. 

PerDayStats = df.groupby(['Trip_Year', 'Trip_Month','Trip_Day','gender','usertype']).agg({'tripduration': ['count', 'mean']})
PerDayStats.columns = ['trip_count', 'avg_trip_length (min)']
PerDayStats = PerDayStats.reset_index()
PerDayStats["avg_trip_length (min)"] = PerDayStats["avg_trip_length (min)"] / 60
PerDayStats.drop(PerDayStats.loc[PerDayStats['gender']==0].index, inplace=True)
PerDayStats
```

```python
# Check the only meaningful data is included

checkgender = PerDayStats['gender'].unique()
print(checkgender)

checkusertype = PerDayStats['usertype'].unique()
print(checkusertype)
```

```python
#FILE 2 - Popular Start Locations in April 2020
startroute = df[['Trip_Year', 'Trip_Month','Trip_Day','start station name', 'start station latitude', 'start station longitude', 'end station id', 'end station name', 'end station latitude', 'end station longitude','tripduration']]
PopularStartRoutes = startroute.groupby(['Trip_Year', 'Trip_Month','start station name', 'start station latitude', 'start station longitude']).agg({'tripduration': ['count']}).reset_index(drop=False)
PopularStartRoutes.columns = ['Trip_Year', 'Trip_Month','start station name', 'start station latitude', 'start station longitude','trip_count']
PopularStartRoutes.loc[:,'trip_count'] = PopularStartRoutes['trip_count'].max()
PopularStartRoutes
```

```python
#FILE 3 - Popular Start Locations in April 2020
endroute = df[['Trip_Year', 'Trip_Month','Trip_Day','start station name', 'start station latitude', 'start station longitude', 'end station id', 'end station name', 'end station latitude', 'end station longitude','tripduration']]
PopularEndRoutes = endroute.groupby(['Trip_Year', 'Trip_Month','end station name', 'end station latitude', 'end station longitude']).agg({'tripduration': ['count']}).reset_index(drop=False)
PopularEndRoutes.columns = ['Trip_Year', 'Trip_Month','end station name', 'end station latitude', 'end station longitude','trip_count']
PopularEndRoutes.loc[:,'trip_count'] = PopularEndRoutes['trip_count'].max()
PopularEndRoutes
```

# Change the file name and save files 1,2,3 for each month in the "Cleaned_Data" folder

```python
PerDayStats.to_csv(r'..\Cleaned_Data\2020_Apr_PerDayStats.csv', index = False)
PopularStartRoutes.to_csv(r'..\Cleaned_Data\2020_Apr_PopularStartRoutes.csv', index = False)
PopularEndRoutes.to_csv(r'..\Cleaned_Data\2020_Apr_PopularEndRoutes.csv', index = False)
```
